```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.linalg
import math
```

```python
#for code reproducibility
np.random.seed(42)

n = 10
#define matrix A
A = np.random.rand(n,n)

#x_true
x_t = np.ones(n)

#compute b
b = A @ x_t
#We try to solve the classical Ax = b
# where now x is the unknown. We need to estimate how much our algorithm is accurate

#the condition number in 2 norm
cond2=np.linalg.cond(A,2)

#the condition number in inf norm
condInf = np.linalg.cond(A,'f')

#solve the linear system
x = np.linalg.solve(A,b)

#relative error
error =  np.linalg.norm(x-x_t) / np.linalg.norm(x_t)

plt.plot(error,'*')
plt.figure()
plt.plot(cond2,'*')
plt.plot(condInf,'*')
plt.legend(['cond 2','cond inf'])
plt.show()
```

```python
cond2  = []
condInf = []
errors = []
size = range(10,101,10)
for i in size:
    A = np.random.rand(i,i)
    x_t = np.ones(i)
    b = A @ x_t
    cond2.append(np.linalg.cond(A,2))
    condInf.append(np.linalg.cond(A,'f'))
    x = np.linalg.solve(A,b)
    errors.append( np.linalg.norm(x-x_t) / np.linalg.norm(x_t))


plt.figure(figsize=(12, 5))
plt.subplot(1,2,1)
plt.plot(size,errors)
plt.xlabel("Size")
plt.ylabel("error")
plt.subplot(1,2,2)
plt.plot(size,cond2)
plt.plot(size,condInf)
plt.legend(['cond 2','cond inf'])
plt.xlabel("Size")
plt.ylabel("condition")
plt.show()
```

#Floating Point Arithmetic

```python
#solution for getting epsilon in the exercise
eps = np.asarray([1],np.float32)

while 1 + eps > 1:
    eps = eps / 2
print(f'The machine epsilon for single precision is: {2*eps}')

eps = np.array([1])

while 1 + eps > 1:
    eps = eps / 2
print(f'The machine epsilon for double precision is: {2*eps}')
```

```python
N=np.arange(1,1e8,500)

e_=np.power((1/N)+1,N)
plt.title('Approx of e')
plt.plot(N,e_)
plt.xlabel("N")
plt.ylabel("e_approx")
print(f"True e: {math.e}")
print(f"Approx e: {e_[-1]}")
plt.figure()

plt.title('Error wrt to real e in log scale')
plt.yscale("log")
plt.grid()
err=math.e-e_
print(math.e - e_[-1])
plt.plot(N,err)
```
